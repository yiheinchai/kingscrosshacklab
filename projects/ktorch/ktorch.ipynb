```python
class Linear():
    def __init__(self, in_features, out_features, bias=True):
        W = torch.randn(in_features, out_features, requires_grad=True)
        B = torch.zeros(out_features, requires_grad=True)

        self.W = W

        if bias:
            self.B = B
            self.parameters = [W, B]
        else:
            self.parameters = [W]

    def __call__(self, x):

        for p in self.parameters:
            p.grad = None

        self.out = (x @ self.W) + self.B

        return self.out
    

class BatchNorm():
    self 
```
